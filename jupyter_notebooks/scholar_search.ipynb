```python
# import necessary libraries
from scholarly import scholarly
```

# Fetching "ALL" details using author name

```python
query = input("Enter author name: ")
search_query = scholarly.search_author(query)
author = next(search_query)
scholarly.fill(author, sections=[])
```

# Search for an author by the id visible in the url of an Authors profile

```python
id = input("Enter author ID: ")
author = scholarly.search_author_id(id)
scholarly.pprint(author)
```

# Search for authors using organisation name

```python
insti = input("Enter name of institute: ")
insti_id = int(scholarly.search_org(insti)[0]['id'])
print(scholarly.search_org(insti))
```

```python
search_query = scholarly.search_author_by_organization(insti_id)
author = next(search_query)
scholarly.pprint(author)
```

# Search by keyword and return a generator of Author objects

```python
keyword = input("Enter keyword: ")
search_query = scholarly.search_keyword(keyword)
scholarly.pprint(next(search_query))
```

# Export a publication to Bibtex by using the `bibtex` property

```python
name = input("Enter paper title: ")
query = scholarly.search_pubs(name)
pub = next(query)
scholarly.bibtex(pub)
```
