
# Animated 3D random walk

Output generated via `matplotlib.animation.Animation.to_jshtml`.


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation as animation

num_particles = int(input("Enter the number of particles: "))
num_steps = int(input("Enter the number of steps: "))
step_length = float(input("Enter the step length: "))
fps = int(input("Enter the frames per second (fps): "))
dpi = int(input("Enter the DPI (dots per inch): "))
axis_length = num_steps * step_length

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.grid(True)
ax.set_axis_on()

# Initialize particle positions
positions = np.zeros((num_particles, 3))
paths = [np.zeros((num_steps+1, 3)) for _ in range(num_particles)]

# Initialize plot lines for paths
path_lines = [ax.plot([], [], [], '-', linewidth=1)[0] for _ in range(num_particles)]

ax.set(xlim3d=(0, 5), xlabel='X')
ax.set(ylim3d=(-5,0), ylabel='Y')
ax.set(zlim3d=(0, 5), zlabel='Z')

path_colors = np.random.rand(num_particles, 3)

# Update function for animation
def update(num):
    global positions
    positions += np.random.randn(num_particles, 3) * step_length
    for i in range(num_particles):
        paths[i][num] = positions[i]
        path_lines[i].set_data(paths[i][:num+1, 0], paths[i][:num+1, 1])
        path_lines[i].set_3d_properties(paths[i][:num+1, 2])
        path_lines[i].set_color(path_colors[i])

ani = animation.FuncAnimation(fig, update, frames=num_steps, blit=False, repeat=False)

filename = "random_walk_paths_animation.mp4"
ani.save(filename, writer='ffmpeg', fps=fps, dpi=dpi)

print(f"Animation saved as {filename}")

plt.show()
```
